What you know: The ratio of the area of the circle to the area of the square is Ï€ / 4.

```
def square(l, ax):
  ax.plot([-l, l], [-l, -l], 'k-')
  ax.plot([-l, l], [l, l], 'k-')
  ax.plot([-l, -l], [-l, l], 'k-')
  ax.plot([l, l], [-l, l], 'k-')
  #plot of a square
```

```
def circle(l, ax):
  theta = np.linspace(0, 360, 100) #100 numbers from 0 to 360
  theta *= u.degree
  x = l * np.cos(theta.to(u.rad))
  y = l * np.sin(theta.to(u.rad))
  ax.plot(x, y, 'k-')
```
