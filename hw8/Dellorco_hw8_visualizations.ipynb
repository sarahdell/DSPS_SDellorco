**Good Visualization**

![1](https://raw.githubusercontent.com/sarahdell/DSPS_SDellorco/master/hw8/GoodVisualizationScreenshot.png)

This image from "Data Dialogue with ChatGPT: Using Code Interpreter to Simulate and Analyse Experimental Data" shows the resulting graph coming from the graphing code ChatGPT supplied using a specific method of inquiry. In this, ChatGPT was asked to make this graph using data from a task where one must find the spring constant of a spring using a comparison of force on the spring and extension of the spring. From Hooke's Law, the graph should be linear, with the slope being the spring constant. This is a good visualization because from looking at the visualization it is clear what is being compared in the plot, the relationship between the two variables is clear, and the k value, axes, and title are all labeled with units (it is not ambiguous). There is no extra noise going on to distract from the message of the visualization and there are no misconceptions that can be easily taken from the visualization (it is not distracting or distorted). It also follows all of Tufte's rules.

**Bad Visualization**

![2](https://raw.githubusercontent.com/sarahdell/DSPS_SDellorco/master/hw8/BadVisualizationScreenshot.png)

This image from "Responsive Professional Development: A Facilitation Approach for Teachersâ€™ Development in a Physics Teaching Community of Practice" is supposed to represent the Professional Development meetings that happened in the Illinois Physics and Secondary Schools (IPaSS) PD schedule.  The image is ambiguous and violates Tufte's 5th rule "Avoid redundancy in communication: The number of information carrying (variable) dimensions depicted should not exceed the number of dimensions in the data. Graphics must not quote data out of context." The graph is showing variables time period using the x-axis, length of meeting using shape, and type of meeting (virtual, in-person, hybrid) using the shade of the shape. Each point is a different meeting. However, there is additionally different colors (red and yellow) within the visualization where color has no significance in indicating a new variable. If it is being used to specifically highlight certain meetings for some reason, it is not indicated in the figure caption. The image also specifically has sections on the time line indicating "Summer" and "School Year," and while there are ticks showing where one year ends and the next one starts, there is no clear divide between the Summer and School Year within one year. In the first year specifically it is very difficult to tell which meetings happened in the summer and which happened during the school year.

**Bad Visualization**

![3](https://raw.githubusercontent.com/sarahdell/DSPS_SDellorco/master/hw8/OtherBadVisualizationScreenshot.png)

This image from "Improving Learning in Science and Mathematics
with Exploratory and Interactive Computational
Modelling" shows the results of an exit survey after a class to get the students' opinions on the effectiveness of the class. The sample size is 115, and it shows the average response from the students on a range of -3 (completely disagree) to 3 (completely agree) on 14 questions. I think this visualization breaks the "Distortion" rule. Many different spreads of data could yield a certain average response. For example, if all the students responded 0 to a question, the average would be 0. If 50 students responded -3 and 50 students responded 3 to a question, that would also yield an average response of 0. It could also be called ambiguous in some ways because the visualization doesn't have any label within it that states the bar indicates the average response value, so you wouldn't know that without reading the article.

```python
import matplotlib.pyplot as plt
import numpy as np
averages = [np.mean(lst) for lst in lists]

# Create a boxplot for each list
plt.boxplot(lists)
plt.scatter(range(1, len(lists) + 1), averages, color='red', marker='o', label='Average')
plt.plot(range(1, len(lists) + 1), averages, color='red', linestyle='-', marker='o')
plt.xticks(range(1, len(lists) + 1), [f'Q{i + 1}' for i in range(len(lists))])
plt.ylabel('Responses')
plt.xlabel('Questions')
plt.title('Boxplot of Responses with Averages')
plt.legend(loc='upper right', bbox_to_anchor=(1.15, 1))
plt.show()
```

Notice how Q3, Q6, and Q8 all have averages of 0.8, but the variance of the data differs, and the values within the lists are drastically different.
