```python
import matplotlib.pyplot as plt
import numpy as np
averages = [np.mean(lst) for lst in lists]

# Create a boxplot for each list
plt.boxplot(lists)
plt.scatter(range(1, len(lists) + 1), averages, color='red', marker='o', label='Average')
plt.plot(range(1, len(lists) + 1), averages, color='red', linestyle='-', marker='o')
plt.xticks(range(1, len(lists) + 1), [f'Q{i + 1}' for i in range(len(lists))])
plt.ylabel('Responses')
plt.xlabel('Questions')
plt.title('Boxplot of Responses with Averages')
plt.legend(loc='upper right', bbox_to_anchor=(1.15, 1))
plt.show()
```
